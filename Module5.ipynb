```python
#1.1 Barplot:
import numpy as np
from matplotlib import pyplot as plt

# creating the dataset
data = {'C':20, 
        'C++':15, 
        'Java':30,
        'Python':35
        }
courses = list(data.keys())
values = list(data.values())

plt.figure(figsize = (10, 5))    #Creates figure with width=7inch and height=5inch

# creating the bar plot
plt.bar(courses, values, color ='violet', width = 0.5)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
#1.2 Lineplot:

# importing matplotlib module
from matplotlib import pyplot as plt    

# x-axis values
x = [2, 4, 6, 8, 10]

# Y-axis values
y = [4, 8, 12, 16, 20]

# Function to plot
plt.plot(x, y)

# function to show the plot
plt.show()
```

```python
#1.3 Histogram:
import matplotlib.pyplot as plt

data=[5, 7, 6, 8, 7, 7, 8]
print(data)

plt.hist(data, color="lightblue", edgecolor="black")
plt.title('HISTOGRAM')
plt.xlabel('Python')
plt.ylabel('NO...')
plt.show()
```

```python
#2.Adding text to graph

#Adding Mathematical Equations as Text Inside the Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
y = x**2

#adding text inside the plot
plt.text(2.5, 80, 'Parabola $Y = x^2$', fontsize = 22)

plt.plot(x, y)

plt.xlabel("X-axis", fontsize = 15)
plt.ylabel("Y-axis",fontsize = 15)

plt.show()
```

```python
#Matplotlib Text Using Rectangular box Around the Text

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10, 0.01)
y = x**2

plt.xlabel("X-axis", fontsize = 15)
plt.ylabel("Y-axis",fontsize = 15)

#Adding text inside a rectangular box by using the keyword 'bbox'
plt.text(-5, 60, 'Parabola $Y = x^2$', fontsize = 22,
         bbox = dict(facecolor = 'yellow', alpha = 0.41))     #bounding box

plt.plot(x, y)

plt.show()
```

```python
#Add Text “Sine wave” on a Figure in Matplotlib

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x,y)

plt.text(3.5, 0.9, 'Sine wave', fontsize = 23)

plt.xlabel('X-axis', fontsize = 15)
plt.ylabel('Y-axis', fontsize = 15)

#plt.grid(True, which='both')
plt.show()
```

```python
#Add Text on a Figure in Matplotlib Using Annotation

import matplotlib.pyplot as plt
import numpy as np

x = ['Rani', 'Meena', 'Raju', 'Jhansi', 'Ram']
y = [5, 7, 9, 2, 6]

plt.bar(x,y)

plt.text(3, 7, 'Student Marks',
         fontsize = 18, color = 'g')

plt.xlabel('Students', fontsize = 15)
plt.ylabel('Marks', fontsize = 15)

plt.annotate('Highest scored', xy = (2.4, 8),
             fontsize = 16, xytext = (3, 9),
             arrowprops = dict(facecolor = 'red'),
             color = 'g')

plt.show()
```

3.Getting and Setting values:Patches

```python
#3.Getting and Setting values:Patches

#Creating Rectangle using patch
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, ax=plt.subplots()
ax.axis([0,20,0,20])
shape=Rectangle((5,5),width=20,height=10,angle=90,color="violet")            #Rectangle((xy,width,height,angle,color)
ax.add_patch(shape)
plt.show()
```

```python
#Creating Circle using patch
import matplotlib.pyplot as plt

from matplotlib.patches import Circle
fig, ax=plt.subplots()
ax.axis([0,20,0,20])
shape=Circle((10,10), radius=5)            #Circle((xy,radius)
ax.add_patch(shape)
plt.show()
```

```python
#Creating Ellipse using patch
import matplotlib.pyplot as plt

from matplotlib.patches import Ellipse
fig, ax=plt.subplots()
ax.axis([0,20,0,20])
shape=Ellipse((10,10), width=10,height=5)            #Circle((xy,height, width)
ax.add_patch(shape)
plt.show()
```

```python
#Creating Regular Polygon using patch
import matplotlib.pyplot as plt

from matplotlib.patches import RegularPolygon
fig, ax=plt.subplots()
ax.axis([0,20,0,20])
shape=RegularPolygon((10,10),numVertices=6,radius=4,facecolor="purple",edgecolor="red")            #numVertices=3,4,5.... etc
ax.add_patch(shape)
plt.show()
```

```python
#Creating Polygon using patch
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

fig, ax=plt.subplots()
ax.axis([0,20,0,20])
array=np.array(((2,2),(2,8),(6,10),(10,8),(10,2))) #Polygon plotting
shape=Polygon(array,color="blue")
ax.add_patch(shape)
plt.show()
```

```python
#5.Time series analysis with Pandas
#Consider Maruti Dataset to show Time Series Analysis with Pandas
import pandas as pd
from google.colab import files
uploaded=files.upload()
print(uploaded)
df = pd.read_csv("MARUTI_1.csv")
df.head()
```

```python
#limit the number of columns to keep things simple.
data = df[['Date','Open','High','Low','Close','Volume','VWAP']]
print(data)
```

```python
#Datetime objects in Python

data.info()
```

```python
#It appears that the Date column is being treated as a string rather than as dates.
#To fix this, we’ll use the pandas to_datetime() feature which converts the arguments to dates.

# Convert string to datetime64
data['Date'] = data['Date'].apply(pd.to_datetime)
data.set_index('Date',inplace=True)
data.head()
```

```python
#Visualising the Time Series data
import matplotlib.pyplot as plt
data['VWAP'].plot(figsize=(10,6),title='Maruti Stock Prices')
plt.ylabel('VWAP')
```

```python
#Subsetting Data Using Pandas Dataframes
       #Instead of working with the entire data,
       #it is also possible to slice the Time Series data to highlight the portion of the data we are interested in.
       #Since the volume weighted average price (VWAP) is a trading benchmark,
       #we shall limit our analysis to only that column.
df_vwap = df[['Date','VWAP']]
df_vwap['Date'] = df_vwap['Date'].apply(pd.to_datetime)
df_vwap.set_index("Date", inplace = True)
df_vwap.head()
```

```python
import pandas as pd

# Create a sample time series data
data = {
    'date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'value': [10, 15, 20, 25, 30]
}

# Convert the 'date' column to datetime format
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index
df.set_index('date', inplace=True)

# Display the DataFrame
print("Sample Time Series Data:")
print(df)
print()

# Accessing specific time periods
print("Accessing specific time periods:")
print("Data for January 2024:")
print(df.loc['2024-01'])  
print()

# Resampling
print("Resampling - Weekly:")
weekly_data = df.resample('W').mean()  # Resample to weekly frequency, taking the mean
print(weekly_data)
print()

# Rolling window calculation
print("Rolling Window Calculation - 3-day moving average:")
rolling_avg = df.rolling(window=3).mean()  # Calculate the 3-day moving average
print(rolling_avg)
print()

# Plotting
import matplotlib.pyplot as plt

# Plot the original data
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['value'], label='Original Data')

# Plot the rolling average
plt.plot(rolling_avg.index, rolling_avg['value'], label='3-day Moving Average', linestyle='--')

plt.title('Time Series Analysis')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
