```python
import seaborn as sns
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]

sns.lineplot(x=x,y=y)

plt.title('Line plot')
plt.xlabel('X-aixs')
plt.ylabel('y-axis')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data={'Group A':[10,15,20,25,30],
      'Group B':[5,10,15,20,25],
      'Group C':[8,12,16,20,24]
      }

sns.boxplot(data=data)

plt.title('Box plot')
plt.xlabel('Groups')
plt.ylabel('values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x=['Category 1','Category 2','Category 3','Category 4']
y=[10,15,20,25]

sns.pointplot(x=x,y=y)

plt.title('Point plot')
plt.xlabel('categorie')
plt.ylabel('values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data=['A','B','A','C','B','A','A','C','C','B']

sns.countplot(x=data)

plt.title('Count plot')
plt.xlabel('categories')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data={'Group A':[10,15,20,25,30],
      'Group B':[5,10,15,20,25],
      'Group C':[8,12,16,20,24]
      }

sns.violinplot(data=list(data.values()))

plt.title('Violin plot')
plt.xlabel('Groups')
plt.ylabel('values')
plt.show()
```
